```python
import pandas as pd
url="/home/shafeen/Documents/My-all-programs--/Probability&&Statistics with python/stat-env/nhanes_2015_2016.csv"
df=pd.read_csv(url)
df=pd.DataFrame(df)
df
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
wgt=df['BMXWT']
wgt
```

```python
plt.figure(figsize=(20,5))
x=range(df['BMXWT'].size)
y=wgt
plt.scatter(x,y,marker='x',s=0.5)
plt.title("Weight of every Individual")
plt.xlabel("Weight")
plt.ylabel("Frequency")
p1=(0,75)
p2=(6000,75)
x_c, y_c=zip(p1,p2)
plt.plot(x_c,y_c,color='red',linewidth=0.75)
plt.show()
```

```python
wgt.mean()
```

```python
plt.figure(figsize=(20,5))
x=range(df['BMXWT'].size)
y=df['BMXWT']
plt.scatter(x,y,marker='*',s=1)
plt.title("Weight of every Individual")
plt.xlabel("Weight")
plt.ylabel("Frequency")
#the mean i think
x_c, y_c=([0,6000],[75,75])
plt.plot(x_c,y_c,color='red',linewidth=0.85)
#the mean the function provided
x_c, y_c=([0,6000],[wgt.mean(),wgt.mean()])
plt.plot(x_c,y_c,color='green',linewidth=2)
plt.show()
```
