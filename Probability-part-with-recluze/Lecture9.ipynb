```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline 
# np.random.seed(0)
num=np.random.uniform(low=0.0,high=1.0)
num
np.random.seed(0)
def flip(num = 1):
    toss=[]
    for i in range(num):
        num=np.random.uniform(low=0.0,high=1.0)
        if num > 0.5:
            toss.append('T')
        else:
            toss.append('H')
    return toss
flip()
flips=flip(10)
flips
values, counts=np.unique(flips,return_counts=True)
values ,counts
```

```python
# flips=['H']
```

```python
from collections import Counter , defaultdict

def get_freqs(flips):
    keys=Counter(flips).keys()
    vals=Counter(flips).values()
    # print (keys)
    # print(vals)
    # return dict(zip(keys,vals))  #there is an error in this if there is no value to the key then it will give error
    
    return defaultdict(int , zip(keys,vals)) #this will return 0 as the default value of int if it cannot find a value or its key
    
freq=get_freqs(flips)
print(freq)
```

```python
prob_h=freq['H']/len(flips)
print(prob_h)
```

```python
#1 to N flips
maximum_flips = 1000

probs = [] 
for num_flips in range(1, maximum_flips):
    flips = flip(num_flips)
    freqs = get_freqs(flips)
    prob_h = freqs['H'] / len(flips)

    probs.append(prob_h)
    
# print(probs)
```

```python
plt.plot(probs)
plt.xlabel("Frequency")
plt.ylabel("Probability")
plt.title("For 1000 flips")
sns.despine(offset=10,trim=True)
plt.show()
```

```python
plt.plot(probs[:10])
plt.xlabel("Frequency")
plt.ylabel("Probability")
plt.title("For 1000 flips")
sns.despine(offset=10,trim=True)
plt.show()
```
